A numpy array is a (list-like) structure that contains numeric data in a row-and-column format.

Numpy is a tool that helps you effectively process multidimensional arrays. (faster than list)

# **Speed test for numpy**

```python
# speed test lists vs array
from timeit import default_timer as timer

a = np.random.randn(1000)  # randn(): return samples from the "standard normal" distribution.
b = np.random.randn(1000)

A = list(a) # converting the numpy array to list
B = list(b) # converting the numpy array to list

print(a.shape)
print(b.shape)

print("*"*50)

print(A)
print(B)
```

```python
def dot_calculation1(lst1,lst2):
    dot = 0
    for i in range(len(lst1)):
        dot += lst1[i]*lst2[i]
    return dot

def dot_calculation2(a,b):
    return np.dot(a,b) # one line
  
iter = 1000 # calculate 1000 times
```

```python
start = timer()
for t in range(iter):
    dot_calculation1(A,B)
end = timer()
t1 = end-start

start = timer()
for t in range(iter):
    dot_calculation2(a,b)
end = timer()
t2 = end-start

print('Time with lists:', t1)
print('Time with array:', t2)
print('Ratio:', t1/t2)
```

# **Dot product**

Calculating the dot product of vectors is important in machine learning tasks.

What is a dot product?

Two equal-length sequences of numbers are multiplied and summed. 

It returns a single number.

[1,2,3] * [4,5,6] = (1 \* 4) + (2 \* 5) + (3 \* 6) = 4 + 10 + 18 = 32

## **Using a list**

```python
lst1 = [1,2,3]
lst2 = [4,5,6]

dot = 0
for i in range(len(lst1)):
  dot += lst1[i] * lst2[i]

print(dot) # However, this is not an intuitive way to do.
```

## **Using numpy**

```python
a = np.array([1,2,3])
b = np.array([4,5,6])
```

```python
# another way to perform the dot product
dot_sum3 = a @ b #
print(dot_sum3)
```

```python
dot_sum2 = (a*b).sum()
print(dot_sum2)
```

```python
# Another way to perform the dot product (manually)
dot = a * b
print(dot)
```

```python
dot_sum = np.sum(dot)
print(dot_sum)
```

```python
# Easy way to calcuate the dot product
# Less code
dot_result = np.dot(a,b)
print(dot_result)
```

# **Linear algebra**

Eigenvalues, Eigenvector: Needed when you perform PCA (principal component analysis) for machine learning.

PCA is used to reduce the dimension of your data. You will learn about this in coming semesters.

Eigenvector does not change direction in a transformation.

Matrix \* eigenvector = Eigenvalue \* eigenvector


```python
# eigenvalues
array1 = np.array([[1,2], [3,4]])
eigenvalues, eigenvectors = np.linalg.eig(array1)
```

```python
print(eigenvalues) # 2 eigenvalues, because we have 2 columns
print(eigenvectors) # column vectors
```

```python
print(eigenvectors[:,0]) # column 0 corresponding to eigenvalue[0]
print(eigenvectors[:,1]) # column 1 corresponding to eigenvalue[1]
```

```python
# verify: Eigenv vector * Eigenvalue = A * Eigen vector
left= eigenvectors[:,0] * eigenvalues[0]
right = np.dot(array1,eigenvectors[:, 0])
print(left)
print(right)
```

```python
# Returns True if two arrays are element-wise equal within a tolerance.
# use allclose()
print(np.allclose(left,right))
```
