```python
# customí•œ ë°©ì‹ìœ¼ë¡œ ë¶ˆìš©ì–´ ì„¤ì •
stopwords = "ì œ ì´ ìˆ í•˜ ê²ƒ ë“¤ ê·¸ ë˜ ìˆ˜ ì´ ë³´ ì•Š ì—† ë‚˜ ì‚¬ëŒ ì£¼ ì•„ë‹ˆ ë“± ê°™ ìš°ë¦¬ ë•Œ ë…„ ê°€ í•œ ì§€ ëŒ€í•˜ ì˜¤ ë§ ì¼ ê·¸ë ‡ ìœ„í•˜ ì € ì „ ë‚œ ì¼ ê±¸ ë­ ì¤„ ë§Œ ê±´ ë¶„ ê°œ ë ì¼ ì´ê±° ë²ˆ ì¤‘ ë“¯ ë•Œ ê²Œ ë‚´ ë§ ë‚˜ ìˆ˜ ê±° ì  ê²ƒ ì˜ ê°€ ì´ ì€ ë“¤ ëŠ” ì¢€ ì˜ ê± ê³¼ ë„ ë¥¼ ìœ¼ë¡œ ì ì— ì™€ í•œ í•˜ë‹¤ ì„ ì•„ ê·¸"
stopwords = list(set(stopwords.split(" ")))
print(stopwords)
```

```python
# ë‹¤ì‹œ ìƒˆë¡­ê²Œ ì¶”ì¶œ 
nouns_new = []
for review in nouns:
  temp = []
  for element in review:
    if (element not in stopwords): # and len(element) > 1: # ë¶ˆìš©ì–´ì™€ ì œê±°
      temp.append(element)
  nouns_new.append(temp)
```

```python
from gensim import corpora # í† í”½ëª¨ë¸ë§
kr_dictionary = corpora.Dictionary(nouns_new) # ë‹¨ì–´ë“¤ì˜ ì‚¬ì „ ë§Œë“¤ê¸° (ì •ìˆ˜ ì¸ì½”ë”©)
print(kr_dictionary)
# ì¶œí˜„ë¹ˆë„ê°€ ì ê±°ë‚˜ ìì£¼ ë“±ì¥í•˜ëŠ” ë‹¨ì–´ëŠ” ì œê±° 
kr_dictionary.filter_extremes(no_below=5, no_above=0.20)

corpus = [kr_dictionary.doc2bow(text) for text in texts] # Term Document Frequency ë§Œë“¤ê¸°, ë‹¨ì–´ê°€ í•´ë‹¹ ë¬¸ì„œì—ì„œ ëª‡ ë²ˆ ì¶œí˜„í•˜ëŠ”ì§€ ì—¬ë¶€
print(corpus[0]) # ìˆ˜í–‰ëœ ê²°ê³¼ì—ì„œ ì²«ë²ˆì§¸ ë‰´ìŠ¤ ì¶œë ¥. ì²«ë²ˆì§¸ ë¬¸ì„œì˜ ì¸ë±ìŠ¤ëŠ” 0
```

```python
# LDA ëª¨ë¸ ìƒì„± - ëŒ€ëµ 
lda_model = gensim.models.ldamodel.LdaModel(corpus=corpus, 
                                           num_topics = 10, 
                                           id2word=kr_dictionary,
                                           random_state=2022,
                                           passes=5,
                                           iterations=50
                                           )
```

```python
topic_table = pd.DataFrame()
```

```python
# ê° ë¬¸ì„œì— ëŒ€í•´ì„œ ë¹„ì¤‘ì´ ë†’ì€ í† í”½ìˆœìœ¼ë¡œ í† í”½ì„ ì •ë ¬í•œë‹¤.
for i, topic_list in enumerate(lda_model[corpus]):
  ## lda_model.per_word_topicsê°€ Trueë©´ topic_list[0]ì„ fasleë©´ topic_list ì „ì²´ë¥¼ í• ë‹¹
  doc = topic_list[0] if lda_model.per_word_topics else topic_list
  #print(doc)           
  doc = sorted(doc, key=lambda x: (x[1]), reverse=True)
  #print(doc)
  for j, (topic_num, prop_topic) in enumerate(doc): # ëª‡ ë²ˆ í† í”½ì¸ì§€ì™€ ë¹„ì¤‘ì„ ë‚˜ëˆ ì„œ ì €ì¥
    if j == 0:
      ## ì •ë ¬ì„ í•œ ìƒíƒœì´ë¯€ë¡œ ê°€ì¥ ì•ì´ ë¹„ì¤‘ì´ ë†’ìŒ
      topic_table = topic_table.append(pd.Series([int(topic_num), round(prop_topic, 4), topic_list]), ignore_index = True)
    else:
      break
```

```python
coherence_values = []
perplexities=[] # ë‚®ì„ ìˆ˜ë¡ ë” ì¢‹ì€ ê°’
model_list = []
for i in range(2,21):
  ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics = i, id2word=kr_dictionary, passes=5,random_state=2022, iterations=50)
  model_list.append(ldamodel)
  coherencemodel = CoherenceModel(model=ldamodel, texts=texts, dictionary=kr_dictionary, coherence='c_v')
  coherence_values.append(coherencemodel.get_coherence())
  perplexities.append(ldamodel.log_perplexity(corpus))
```
