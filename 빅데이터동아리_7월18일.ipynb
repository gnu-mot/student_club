```python
# Box Plot
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(8,6))
sns.boxplot(x= df['fare'])
```

```python
# IQR Íµ¨ÌïòÍ∏∞
# quantile , Î∞±Î∂ÑÏúÑÏàò Íµ¨ÌïòÍ∏∞ (Ï†ïÌôïÌûàÎßêÌïòÎ©¥ 4Î∂ÑÏúÑÏàò)
Q1 = df['fare'].quantile(.25)
Q3 = df['fare'].quantile(.75)
```

```python
IQR = Q3 - Q1
```

```python
lower_bound = Q1 - (1.5 * IQR)
upper_bound = Q3 + (1.5 * IQR)
print(lower_bound)
print(upper_bound)
```

```python
print(df['fare'].min())
```

```python
print(df['fare'].max())
```

```python
outlier = df[(df['fare']<lower_bound) | (df['fare']>upper_bound)].reset_index(drop=False)
```

```python
outlier['fare'].min()
```

```python
len(outlier)
```

```python
# Ïù¥ÏÉÅÏπò Îç∞Ïù¥ÌÑ∞ÏóêÏÑú Ïó¨ÏÑ± Ïàò Íµ¨ÌïòÍ∏∞
print(len(outlier[outlier['sex'] == 'female']))
```
